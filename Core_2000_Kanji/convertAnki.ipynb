```python
import pandas as pd
import os 
```

```python
# Specify the file path
file_path = os.path.join(os.getcwd(), "Core_2000.txt")

print(file_path)
column_headers = ["ID",   "term",  "term_hiragana",  "hiragana",  "meaning", "space1", "part-of-speech", "space2", "sentence", "hiragana_sentence",   "kana_sentence",  "romanized_sentence",  "blank_sentence", "space3", "core2000_identifier", "id1", "id2", "id3"]
# Read the file using `read_table`
df = pd.read_table(file_path, sep="\t", encoding="utf-8",names=column_headers)
```

```python
df.head()
```

```python
export_df = df.loc[:,['term', 'term_hiragana', 'meaning']]
export_df['answer'] = export_df['term_hiragana'] + " \t " + export_df['meaning']
```

```python
export_df.head(5)
```

```python
chunk_size = 10

# Loop through the DataFrame in chunks
for i in range(0, len(export_df), chunk_size):
    # Get the current chunk
    chunk = export_df[['term', 'answer']].iloc[i:i+chunk_size]
    
    # Define the filename with chunk number
    filename = f"Kanji/kanji_day_{(i+1)//10}.txt"
    
    # Save the chunk to a separate text file
    chunk.to_csv(filename, header=False, index=False, encoding='utf-8')
```
